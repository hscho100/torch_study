```python
collection = [(filename,kobill.open(filename).read()) for filename in kobill.fileids()]
```

```python
globalLexicon = dict() 
globalDocument = list()
globalPosting = list()
weightPosting = list()
documentLength = [0 for _ in range(len(collection))]
globalMaxTF = dict()
globalTotalTF = dict()

for docName, docContent in collection:
    docIdx = len(globalDocument)
    globalDocument.append(docName)
    localPostring = defaultdict(int)
    for term in [_ for token in word_tokenize(docContent) for _ in ma.pos(token)]:
        if len(term[0]) > 1 and term[1].startswith("N"):
            localPostring[term] += 1
    globalMaxTF[docIdx] = max(localPostring.values())
    globalTotalTF[docIdx] = sum(localPostring.values()) # tf 값 구할 때, 분모로 sum값 쓸 수도 있음
################################################## local 
    for term, freq in localPostring.items():
        if term in globalLexicon.keys():
            termIdx = list(globalLexicon.keys()).index(term)
            postingIdx = len(globalPosting)
            globalPosting.append([docIdx,freq,globalLexicon[term]])
            globalLexicon[term] = postingIdx
        else:
            termIdx = len(globalLexicon.keys())
            postingIdx = len(globalPosting)
            globalLexicon[term] = postingIdx
            globalPosting.append([docIdx, freq, -1])
```

```python
def idf(df,N):
    return log10(N/df) # 이거 쓰면 불용어(to, a, the)는 무조건 사라짐, 그래도 많이 사용 됨!

def smoothingIdf(df,N):
    return log10(N/(1+df))

def probabilisticIDF(df,M):
    return log10((N-df+1)/df)
```

### with Eucldiean distance

```python
searchResult = defaultdict(float)
```

```python
for term, postingIdx in globalLexicon.items():
    while 1:
        if postingIdx == -1:
            break
        new = globalPosting[postingIdx]
        searchResult[globalDocument[new[0]]] += (new[1] - QWM[term])**2 
        postingIdx = new[2]
for doc, distance in searchResult.items():
    searchResult[doc] = sqrt(distance)
```
